```python
#Line Plot using Matplotlib

#Import Libraries
import matplotlib.pyplot as plt
import numpy as np

# Generate some data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a line plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Sine Wave', color='blue')
plt.title('Sine Wave')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
```

This line plot showcases a sine wave generated from mathematical data, demonstrating the sine function's periodic behavior.

```python
#Scatter Plot using Matplotlib

#Import Libraries
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
np.random.seed(42)
x = np.random.rand(50)
y = np.random.rand(50)

# Create a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y, label='Random Data', color='red', marker='o')
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
```

The scatter plot illustrates random data points distributed across the chart, displaying no particular pattern.

```python
#Histogram using Seaborn

#Import Libraries
import seaborn as sns
import numpy as np

# Generate random data
np.random.seed(42)
data = np.random.normal(0, 1, 1000)

# Create a histogram
plt.figure(figsize=(8, 6))
sns.histplot(data, kde=True, color='green', bins=30)
plt.title('Histogram with KDE')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

This histogram with a kernel density estimate (KDE) reveals the distribution of random data, showing a bell-shaped curve characteristic of a normal distribution.

```python
#Bar Chart using Seaborn

#Import Libraries
import seaborn as sns
import pandas as pd

# Create sample data as a pandas DataFrame
data = {
    'Category': ['A', 'B', 'C', 'D'],
    'Values': [25, 40, 30, 45]
}
df = pd.DataFrame(data)

# Create a bar chart
plt.figure(figsize=(8, 6))
sns.barplot(data=df, x='Category', y='Values', palette='Blues')
plt.title('Bar Chart')
plt.xlabel('Category')
plt.ylabel('Values')
plt.grid(True)
plt.show()
```

The bar chart visualizes categorical data, comparing values in different categories (A, B, C, D) using bars of varying heights.
